```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# load the data
embeddings = np.load('embeddings.npy')
```

```python
# reduce 
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)
```

```python
# read the csv file for labeling
all_globins = pd.read_csv('all_globins.csv')
#merged_df = pd.read_csv('merged_data.csv')
```

```python
all_globins = all_globins.iloc[0:283]
#merged_df = merged_df.iloc[0:283]
```

```python
labels = all_globins['EntryID']

def get_color(label):
    if "beta" in label:
        return "orange"
    if "myo" in label:
        return "red"
    if "alpha" in label:
        return "purple"
    if "beta" in label:
        return "black"
    else:
        return "yellow"

colors = [get_color(label) for label in labels]

df = pd.DataFrame({
    'x': [e[0] for e in reduced_embeddings],
    'y': [e[1] for e in reduced_embeddings],
    'label': labels,
    'color': colors
})

fig = px.scatter(
    df,
    x='x',
    y='y',
    hover_name='label',
    color='color',
)

# make the plot square 
fig.update_layout(
    autosize=False,  
    width=750,  # set the dimensions here
    height=550,  
    xaxis_scaleanchor='y',  
    yaxis_scaleanchor='x',  
)

fig.show()
```
