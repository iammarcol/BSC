```python
import pandas as pd
import glob

# Specify the directory where your CSV files are located
directory = '/Users/markol/Desktop/frustlact/filescopy/output/'

# Use glob to find all CSV files in the directory
file_paths = glob.glob(directory + '*.txt')

# Initialize an empty list to store DataFrames
dfs = []

# Loop through the file paths and read each CSV file into a DataFrame
for file_path in file_paths:
    df = pd.read_csv(file_path, delimiter='\t')
    # Extract the filename (without the path and extension) and use it as a column level
    file_name = file_path.split('/')[-1].split('.')[0]
    df.columns = pd.MultiIndex.from_product([[file_name], df.columns])
    dfs.append(df)

# Concatenate all DataFrames horizontally into a single DataFrame
combined_df = pd.concat(dfs, axis=1)

combined_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

os.makedirs("plots_oct_9", exist_ok=True)

# Replace with actual pseudotime values

df = pd.DataFrame(combined_df)

# Function to create labels for each amino acid and color them based on frustration
def create_amino_acid_labels(pseudotime, row, amino_acid):
    fig, ax = plt.subplots()
    
    ax.set_xlabel("Pseudotime")
    ax.set_ylabel("Frustration")
    ax.set_title(str(amino_acid))  # Convert the Series to a string
    
    for pt, frustration, aa in zip(pseudotime, row, amino_acid):
        color = 'gray'
        if frustration < -1:
            color = 'red'
        elif frustration > 0.55:
            color = 'green'
        
        ax.annotate(aa, (pt, frustration), textcoords="offset points", xytext=(10, 0), va='center', fontsize=12, color=color)
    
    # Add threshold lines
    threshold_red = -1
    threshold_green = 0.55
    
    ax.axhline(y=threshold_red, color='red', linestyle='--', label='Frustration < -1', linewidth=0.5)
    ax.axhline(y=threshold_green, color='green', linestyle='--', label='Frustration > 0.55', linewidth=0.5)
    
    ax.legend()
    
    return fig, ax

# Iterate through rows and create plots with only amino acid labels and colored based on frustration
for index, row in df.iterrows():
    amino_acid = row.str.split(',').str[0].str.strip('"')  # Extract amino acid names
    frustration_values = row.str.split(',').str[1].str.strip('"').replace('N/A', np.nan).astype(float)
    fig, ax = create_amino_acid_labels(pseudotime_values, frustration_values, amino_acid)
    ax.set_title(f"{index}")
    ax.set_ylim(-4, 4)
    ax.set_xlim(0.5, 18)
    
    # Save each plot
    file_name = f"plots_oct_9/{index}.png"
    plt.savefig(file_name)
    plt.close()
```
